```python
# Definition for singly-linked list.
class ListNode:
  def __init__(self, val=0, next=None):
    self.val = val
    self.next = next
def connectListNode(node,i):
  new_node=ListNode(i)
  node.next=new_node
  return new_node

def createListNode(size):
  head=ListNode()
  current_node=head
  for i in range(size):
    current_node=connectListNode(current_node,i+1)
  return head

def middleNode(head):
  slow=head.next
  fast=head.next
  while fast and fast.next:
    slow=slow.next
    fast=fast.next.next
  return slow.val

Node1=createListNode(5)
print(middleNode(Node1))

Node2=createListNode(6)
print(middleNode(Node2))
```
